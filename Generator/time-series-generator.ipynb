```python
import numpy as np
import numexpr as ne
import pandas as pd
import matplotlib.pyplot as plt
from generator import TSGenerator
```

# Generate by expression

```python
# expr - выражение с параметром t
# time - значения параметра t
# noise_std - стандартное отклонение шума

def GenerateByExpression(expr, time, noise_std = 1):
    ts = list()
    for t in time:
        ts.append(ne.evaluate(expr))
    
    noise = np.random.normal(0, noise_std, len(time))
    ts += noise
    
    return ts
```

# AR

```python
# Параметры для генерации
n = 100  # Длина временного ряда
ar_params = [0.5, -0.3]  # Коэффициенты AR(2)
noise_std = 1  # Стандартное отклонение шума

# Генерация временного ряда
time_series = TSGenerator.generate_ar_series(n, ar_params, noise_std)

# Визуализация временного ряда
plt.figure(figsize=(10, 5))
plt.plot(time_series, color='blue')
plt.title('Сгенерированный временной ряд на основе модели AR')
plt.xlabel('Время')
plt.ylabel('Значение')
plt.grid()
plt.show()
```
